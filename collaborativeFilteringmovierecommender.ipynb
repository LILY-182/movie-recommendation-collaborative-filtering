```python
#read the data set
#users data set

import pandas as pd
users = pd.read_csv('https://msi.martial.co.ke/data/file.csv')
users
```

```python
#Movie IDS dataset
movies = pd.read_csv('https://msi.martial.co.ke/data/Movie_Id_Titles.csv')
movies
```

```python
#merge the users and movies dataset
df = pd.merge(users, movies, on= 'item_id')
df
```

```python
#Exploratory data analysis
#1. information
df.info()
```

```python
#2. max,average and min rating
df['rating'].max()
df['rating'].min()
df['rating'].mean()
df['title'].count()
```

```python
#3.statistical analysis
#total usders = 943
#total movies = 100003
df.describe()
```

```python
#4.checking empties
df.isnull().sum()
```

```python
#recommendation based on
#1. rating
#2. number of users
```

```python
#group movies(categorical- text/object) based on the average rating
mean_ratings = df.groupby('title')['rating'].mean()

mean_ratings.sort_values(ascending = False)
```

```python
#How many users rated a particular movie
rating_count = df.groupby('title')['rating'].count()
rating_count.sort_values(ascending = False)
```

```python
#create a data frame that will contain the title, mean rating, number of views
df_new = pd.DataFrame(mean_ratings)
df_new['number_of_views'] = pd.DataFrame(rating_count)
df_new.sort_values(by='number_of_views', ascending= False)
```

```python
#PIVOT TABLE
#data can be viewed in 3D.
# x = columns
# y = indexing
# z = values
```

```python
pd.set_option('display.max_columns', 100)
pivot = df.pivot_table(index = 'user_id', columns = 'title', values = 'rating')
pivot
```

```python
#selecting a movie from the pivot
selected_movie = pivot['Godfather, The (1972)']
```

```python
#correlation, relationship between 2 variables,
similar_movies = pivot.corrwith(selected_movie)
import warnings
warnings.filterwarnings('ignore')
similar_movies.sort_values(ascending = False)
```

```python
similar_moviesdf = pd.DataFrame(similar_movies, columns = ['correlation'])
similar_moviesdf.sort_values('correlation', ascending = False)
```

```python
#join the similar_moviesdf with df_new ['number of views']
similar_moviesdf = similar_moviesdf.join(df_new['number_of_views'])
similar_moviesdf
```

```python
#reccommendation
#positive correlation
#high number of views
condition = similar_moviesdf['number_of_views'] > 400
similar_df = similar_moviesdf[condition]

recommend_mov = similar_df.sort_values('correlation', ascending = False)
recommend_mov.head(20)
```
